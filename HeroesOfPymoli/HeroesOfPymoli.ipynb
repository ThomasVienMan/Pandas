```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
#print(purchase_data)
```

## Player Count

```python
count_row = purchase_data.shape[0]
print(count_row)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
df = pd.DataFrame(purchase_data)
unique_names = df['Item Name'].unique()
length = len(unique_names)
price = df['Price'].sum()
average_price = price/780
purchases = df['Purchase ID'].count()
purchase_dict = [{'Number of Unigue Items' : length , 'Average Price' : average_price , 'Number of Purchases' : purchases , 'Total Revenue' : price}]
purchase_df = pd.DataFrame(purchase_dict)
purchase_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_count = purchase_data["Gender"].value_counts()
gender_count.head()
gender_data = pd.DataFrame(gender_count)
#gender_data
gender_data['Percentage'] = gender_data['Gender']/gender_data['Gender'].sum()
gender_data
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
df = pd.DataFrame(purchase_data, columns=['Gender','Purchase ID','Price'])
df['Price'].groupby(df['Gender']).sum()
df['Price'].groupby(df['Gender']).mean()
df['Purchase ID'].groupby(df['Gender']).count()
df2 = pd.DataFrame(df['Purchase ID'].groupby(df['Gender']).count())
df2['Average Purchase Price'] = df['Price'].groupby(df['Gender']).mean()
df2['Total Purchase Value'] = df['Price'].groupby(df['Gender']).sum()
df2['Avg Total Purchase per Person'] = df['Purchase ID'].groupby(df['Gender']).count()/df['Price'].groupby(df['Gender']).mean()
df2
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
df = pd.DataFrame(purchase_data, columns=['Age'])

age_ranges = ["[{0} - {1})".format(age, age + 10) for age in range(0, 100, 10)]
age_ranges
count_unique_age_ranges = len(age_ranges)
count_unique_age_ranges
df['Total Count'] = pd.cut(x=df['Age'], bins=count_unique_age_ranges, labels=age_ranges)
df['Total Count'].value_counts()
df_age = pd.DataFrame(df['Total Count'].value_counts())
df_age['Percentage of Players'] = df['Total Count'].value_counts()/780
df_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
df = pd.DataFrame(purchase_data, columns=['SN','Purchase ID','Price'])
df['Purchase ID'].groupby(df['SN']).count()
df['Price'].groupby(df['SN']).sum()
df['Price'].groupby(df['SN']).mean()
df2 = pd.DataFrame()
df2['Purchase Count'] = df['Purchase ID'].groupby(df['SN']).count()
df2['Average Purchase Price'] = df['Price'].groupby(df['SN']).mean()
df2['Total Purchase Value'] = df['Price'].groupby(df['SN']).sum()
sorted_df = df2.sort_values(by='Total Purchase Value', ascending=False)
sorted_df

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
df = pd.DataFrame(purchase_data, columns=['Item ID','Item Name','Purchase ID','Price'])

df2 = pd.DataFrame()
df2['Purchase Count'] = df['Purchase ID'].groupby(df['Item Name']).count()
df2['Average Purchase Price'] = df['Price'].groupby(df['Item Name']).mean()
df2['Total Purchase Value'] = df['Price'].groupby(df['Item Name']).sum()
sorted_df = df2.sort_values(by='Purchase Count', ascending=False)
sorted_df

```
